#### Boxplot between age and heart disease

From this, we can see that there is a high correlation between age and heart disease. There is also a presence of heart disease in individuals between the age of 50 and 70, and higher in individuals between the age of 50 and 70.

#### Boxplot between sex and heart disease

From this, we can see that there is a high correlation between sex and heart disease. There is also a higher presence of heart disease in males1 than females.

#### Boxplot between restecg and heart disease

From this, we can see that there is a high correlation between restecg and heart disease. There is also a higher presence of heart disease in individuals with ST-T wave abnormality than those with normal restecg and those showing probable or definite left ventricular hypertrophy.

#### Boxplot between thal and heart disease

From this, we can see that there is a high correlation between thal and heart disease. There is also a higher presence of heart disease in individuals with reversal defect thal than those with fixed defect thal. There is no presence of heart disease in individuals with normal thal.

#### Crosstab between thal and heart disease

From this crosstab, we can see that the higher the thal value, the higher the heart disease value.

For `thal = 3`, most individuals (127) have no heart disease (heart_disease = 0), while fewer have higher severity levels. For `thal = 6`, the counts are relatively small and evenly spread across heart disease categories. In contrast, `thal = 7` shows a higher number of individuals with heart disease, particularly for `heart_disease = 1, 3, and 4`. This suggests that `thal = 3` is more associated with no heart disease, while `thal = 7` is more associated with heart disease presence.

#### Crosstab between sex and heart disease

From this, we can see that among females (sex = 0), a majority (71) have no heart disease (heart_disease = 0), and fewer fall into higher severity categories. Among males (sex = 1), the number of individuals with no heart disease (89) is slightly higher than in females, but the counts for higher heart disease severity levels (1 to 4) are significantly greater compared to females.

This indicates that males have a higher prevalence of heart disease across all severity levels compared to females.

#### Pairplot between chol and thalach, and visualising heart disease sevierity.

We can see that individuals with no heart disease (0) tend to have higher thalach values (higher heart rates), while severe heart disease cases (14) generally cluster at lower thalach values.

Cholesterol values (chol) are primarily concentrated between 150 and 300 across all heart disease categories, with minimal patterns correlating cholesterol to disease severity.

This suggests that thalach (maximum heart rate) is a stronger indicator of heart disease presence than cholesterol levels.

##### What are the most significant predictors of heart disease?
Gender:
From the crosstab analysis, males (`sex = 1`) have a higher occurrence of heart disease compared to females (`sex = 0`). For instance:
    - 89 males had `heart_disease = 0` (no disease), but a significant number (45, 28, 28, 11) were categorized under heart disease severity `14`.
    - In contrast, females had fewer severe heart disease cases, with most individuals showing no or mild disease.

Age:
From the boxplot, we can see that most individuals with heart diseases are between the ages of 50 and 70 years.


##### Are there notable patterns (e.g., higher risk in certain age groups or genders)
Males appear at a higher risk of heart disease compared to females, with more severe cases recorded in the crosstab. This suggests that gender plays a significant role in heart disease prevalence.

Lower heart rates (`thalach`) are consistently associated with severe heart disease.

In summary, low `thalach` (maximum heart rate), resting ecg, male gender, and age are significant predictors of heart disease, while cholesterol has a less prominent impact.
